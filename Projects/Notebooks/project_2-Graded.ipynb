```python
#Renaming the Column Names
#import re
#1. Replacing the period ('.') in the column names
dbOne = db.rename(columns=lambda x: x.replace('.', ' '), inplace=True)
#2. Replacing the character 'x' from columns
dbOne = db.rename(columns=lambda x: x.replace('x', ''), inplace=True)
#3 Make the first letter as capitals
dbOne = db.rename(columns=lambda x: x.title(), inplace=True)
#4 Replace 'St'
dbOne = db.rename(columns=lambda x: x.replace('St',''), inplace=True)
#5 Replacing 'Nd'
dbOne = db.rename(columns=lambda x: x.replace('Nd',''), inplace=True)
#6 Replacing "Rd"
dbOne = db.rename(columns=lambda x: x.replace('Rd',''), inplace=True)
#6 Replacing 'Th'
dbOne = db.rename(columns=lambda x: x.replace('Th',''), inplace=True)
#7 Replace 'Week'
dbOne = db.rename(columns=lambda x: x.replace('Week',''), inplace=True)
#8 Replace 'Inverted' with blank in place of column 'Artist Inverted'
dbOne = db.rename(columns=lambda x: x.replace('Artist Inverted','Artist'))
#Output with the refined column names
dbOne
```

```python
#Stripping Track column
dbTwo = dbOne
dbOne['Track'] = dbOne['Track'].apply(lambda x: x[ : 20])
dbTwo
```

##### Using Pandas' built in `melt` function, pivot the weekly ranking data to be long rather than wide. As a result, you will have removed the 72 'week' columns and replace it with two: Week and Ranking. There will now be multiple entries for each song, one for each week on the Billboard rankings.

```python
#Get Column Names into a list
dbColumns = list(dbOne)
dbColumns
#Use Column Names list to use first 8 columns as the standard columns and make melt on the Weeks and Rank
dbMelt = pd.melt(dbOne, id_vars = dbColumns[0:7], var_name = 'Week', value_name = 'Rank').dropna()
dbMelt['Rank'] = dbMelt['Rank'].astype(int)
dbMelt['Week'] = dbMelt.Week.astype(int)
dbMelt

#JB great structure!
```

```python
#Sort the Melted DataFrame..
dbMeltSortByTrack = dbMelt.sort_values(['Track','Week'], ascending=True)
#Deleting Year column as it is only year 2000
dbMeltSortByTrack.drop('Year',axis=1,inplace=True)
dbMeltSortByTrack

#JB sort artist first
```

```python
#Show the ranking for a song for an artist per week'
dbPivot = pd.pivot_table(dbMeltSortByTrack, index=['Artist','Track','Rank'])
                        
dbPivot
```

## Step 3: Visualize your data.

##### Using a plotting utility of your choice (Tableau or python modules or both), create visualizations that will provide context to your data.
```python
# Making separate Data Frame
dScatter = dbMeltSortByTrack
# Making subset of Genres 
dScatter['GenreSub'] = np.select([dScatter.Genre == 'Rock', dScatter.Genre == 'Rap', dScatter.Genre == 'Country'],
                        ['Rock','Rap','Country'],'Troubles')
# Fixing colors for each Genre
for color, label in zip('rbgy', ['Rock','Rap','Country','Troubles']):
   GenreSub = dScatter[dScatter.GenreSub == label]
   plt.scatter(GenreSub.Week,GenreSub.Rank, s=25,c=color, label=str(label))
plt.xlabel('Week')
plt.ylabel('Rank')
plt.gca().invert_yaxis()

plt.legend()

#JB what are TROUBLES??? LOL
```

The above plot shows major 3 and other Genres with their weekly ranking

```python
#Group the DataFrame by Genre to find out weekly count
dbGroupBy = dbMeltSortByTrack.groupby("Genre")

#Making Series for Genres and Counts separately
genres = []
counts = []

for name, group in dbGroupBy:
    genres.append( name )
    counts.append( len(group) )

#Regrouping Genres and Counts together into DataFrame    
dfToPlot = pd.DataFrame( {"genre" : genres, "counts": counts } )
#Plotting the barchart
dfToPlot.plot(kind = "bar", x = "genre", y = "counts")
```

The above graph shows Total count of Genres in the Billboard's Top 100. Rock is highest Genre that appeared in Billboard charts and Country is the next best Genre followed by Rap.

```python
dbGrpByArtist = dbMeltSortByTrack[dbMeltSortByTrack.Rank <= 20].groupby("Artist")


artists = []
ranks = []

for name, rank in dbGrpByArtist:
    artists.append( name )
    ranks.append( len(rank) )

dfArtist = pd.DataFrame( { 'Artist' : artists, 'Rank' : ranks } )


dfArtist.plot(kind = "bar", x = "Artist", y = "Rank")
```

The above bar chart shows the artists with songs count below the rank 20 for the year 2000.
