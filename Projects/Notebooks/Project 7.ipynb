#### 5.1 Conduct the PCA

```python
# Create a clean data frame 
# ap1 = ap[['airport','year','departure cancellations','arrival cancellations']]
# print ap1.head()
#ap = df[['airport','year','departure_cancellations','arrival_cancellations']]
```

Preparing the data for PCA analysis.

For PCA analysis we don't need Y axis because it is unsupervised learning method

```python
x = df.ix[:,1:].values
y = df.ix[:,:1].values
# x = ap.ix[:,1:].values
# y = ap.ix[:,:1].values
```

Standardizing the X coordinate

```python
x_standard = StandardScaler().fit_transform(x)
x_standard
```

Create covariance matrix to build Eigen Values and Eigen Vectors.

```python
cov_mat = np.cov(x_standard.T)
print cov_mat
eigenValues, eigenVectors = np.linalg.eig(cov_mat)
```

```python
# Checking the Eigen Vales
print(eigenValues)
```

```python
#print(eigenVectors)
```

```python
# Making Eigan Pairs to combine Eigen Values and Eigen Vectors
eigenPairs = [(np.abs(eigenValues[i]), eigenVectors[:,i]) for i in range(len(eigenValues))]
eigenPairs
```

```python
# Sorting the Eigen Pairs in descending order to identify the farther eigen values on which the PCA model can be built.
eigenPairs.sort()
eigenPairs.reverse()
for i in eigenPairs:
    print(i[0])
```

The top 3 eigen values can be taken for building 3 dimensional PCA representation.

```python
EigenSum = sum(eigenValues)
EigenSum
```

```python
ExpVar = [(i / EigenSum)*100 for i in sorted(eigenValues, reverse=True)]
CumulativeExpVar = np.cumsum(ExpVar)
```

```python
print ExpVar
print CumulativeExpVar
```

```python
# Building two dimensional matrix by reshaping the eigen pairs
Pmatrix = np.hstack((eigenPairs[0][1].reshape(18,1),
                      eigenPairs[1][1].reshape(18,1)))
Pmatrix
```

```python
Y = x_standard.dot(Pmatrix)
```
